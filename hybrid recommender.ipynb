```python
articles_df = pd.read_csv(r'dataset\shared_articles.csv')
articles_df = articles_df[articles_df['eventType'] == 'CONTENT SHARED']
articles_df.head()
```

Contains info about the articles shared in the platform

there are two possible event types at timestamp
- CONTENT SHARED: article was shared in the platform and is available for users
- CONTENT REMOVED: article was removed from the platform and not available for futher recommendation

for simplicity, we only consider 'CONTENT SHARED'

```python
interactions_df = pd.read_csv(r'dataset\users_interactions.csv')
interactions_df.head(10)
```

Contains logs of user interactions on shared articles (it can joined to articles_df by contentid)

The eventType values
- VIEW: user has opened the article
- LIKE: user has liked the article
- COMMENT CREATED: user created a comment in the article
- FOLLOW: user chose to be notified on any new comment in tha article
- BOOKMARK: user has bookmarked the article

### Data muging

As there'r different interactions type, we associate them with weight or strength, assuming that

```python
event_type_strength = {
    'VIEW':1.0,
    'LIKE':2.8,
    'BOOKMARK':2.5,
    'FOLLOW':3.0,
    'COMMENT CREATED':4.0,
}

interactions_df['eventStrength'] = interactions_df['eventType'].apply(lambda x: event_type_strength[x])
```

Recommender system has a problem known as **user cold-star**, for this reason, we are keeping in the dataset only users with at leas 5 interactions.

```python
users_interactions_count_df = interactions_df.groupby(['personId', 'contentId']).size().groupby('personId').size()
print('# users: %d' % len(users_interactions_count_df))
users_with_enough_interactions_df = users_interactions_count_df[users_interactions_count_df >= 5].reset_index()[['personId']]
print('# users with at least 5 interactions: %d' % len(users_with_enough_interactions_df))
```

```python
print('# of interactions: %d' % len(interactions_df))
interactions_from_selected_users_df = interactions_df.merge(users_with_enough_interactions_df, 
               how = 'right',
               left_on = 'personId',
               right_on = 'personId')
print('# of interactions from users with at least 5 interactions: %d' % len(interactions_from_selected_users_df))
```
