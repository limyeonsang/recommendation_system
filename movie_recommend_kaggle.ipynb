# Demographic Filtering

시작하기 전에
- 영화 점수를 매기거나 평점을 매기기 위해 미터법이 필요하다.
- 모든 영화의 점수를 계산한다.
- 점수를 정렬하고 최고 평점의 영화를 사용자에게 추천한다

영화의 평균 평점을 점수로 이용할 수 있다. 하지만 3명의 평가로 평균 평점이 8.9인 영화가 40명의 평가로 평균 평점이 7.8인 영화보다 더 나은 것으로 여겨지기 때문에 이것을 사용하는 것은 불공평하다.

따라서, * IMDB의 weighted rating을 사용한다. \
`*IMDB: 영화, 배우, 드라마 등의 정보를 제공하는 온라인 DB이다.`



- v: 영화의 투표수, 즉 평가자 수
- m: 차트에 나열할 때 필요한 최소 투표수
- R: 영화의 평균 평점
- C: report 전체 투표의 평균

```python
C = df2['vote_average'].mean()
C
```

```python
m = df2['vote_count'].quantile(0.9)
m
```

```python
df2['vote_count'].shape
```

```python
q_movies = df2.copy().loc[df2['vote_count'] >= m]
q_movies.shape
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m) * R) + (m/(m+v) * C) 
```

```python
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
```

```python
q_movies = q_movies.sort_values('score', ascending=False)

q_movies[['title', 'vote_count', 'vote_average', 'score']].head(10)
```

```python
pop = df2.sort_values('popularity', ascending=False)

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))

plt.barh(pop['title'].head(6), pop['popularity'].head(6), align='center', color='skyblue')
plt.gca().invert_yaxis()
plt.xlabel('Popularity')
plt.title('Popular Movies')
```

demographic recommender에서 모든 사용자에게 추천할 수 있는 영화의 일반적인 차트를 제공한다. 하지만 이는 특정 사용자의 관심사와 취향에 맞춰주지 않는다.

# Content Based Filtering

이 영화 추천 시스템에서는 overview, cast, crew, keyword, tagline 등이 다른 영화와의 유사성을 찾는데 사용된다. 그 뒤 비슷할 가능성이 가장 높은 영화가 추천된다.



## Plot description based recommender

overview를 바탕으로 모든 영화의 pairwise similarity점수를 계산하고 이를 바탕으로 영화를 추천한다.

```python
df2['overview'].head()
```

우리는 각 overview의 단어 벡터를 변환할 필요가 있다. 따라서 각 overview에 대한 **Term Frequency-Inverse Document Frequency(TF-IDF)** 벡터를 계산한다.


그렇다면 **Term Frequency**가 무엇일까? 이는 문서에 있는 단어의 상대 빈도수이며 용어는*(instances/total instance)*와 같이 주어진다. \
**Inverse Document Frequency**는 *log(number of docs/docs with term)*로 주어지는 용어가 포함된 문서의 상대 개수이다. \
각 단어가 나타나는 문서에 대한 전체 중요도는  **TF * IDF**와 같다. 

`💡(for more info about TF-IDF, go notion-machine learning-Recommendation Algorithms) `


TF-IDF는 각 열이 overview 어휘의 단어를 나타내고 각 행이 영화를 나타내는 행렬을 제공한다. 이것은 overview에서 자주 등장하는 단어의 중요성을 줄이기 위해 수행된다.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')

df2['overview'] = df2['overview'].fillna('')

tfidf_matrix = tfidf.fit_transform(df2['overview'])

tfidf_matrix.shape
```

4,800개의 영화를 묘사하기 위해 20,000개 이상의 서로 다른 단어들이 사용되었다는 것을 알 수 있다.

이제 이 행렬을 이용해, 유사성 점수를 계산할 수 있다.
유사성 점수를 계산하는 여러 방법들이 있다.
- euclidean
- Pearson
- cosine similarity scores

`💡(for more info about cosine similarity go notion-machine learning-Recommendation Algorithms) `

 **cosine similarity**를 사용해 두 영화의 유사성을 나타내는 숫자 수량을 계산한다. 왜냐하면 cosine similarity는 크기에 관계없이 쉽고 빠르기 때문이다.

수학적으로, 이것은 다음과 같이 정의된다.




TF-IDF vectorizer를 이용했기 때문에, dot product(벡터의 내적)를 계산하면 cosine similarity가 직접 주어진다. 그러므로 더 빠른, sklearn의 **linear_kernel()**을 cosine_similarities()대신 사용한다.

```python
from sklearn.metrics.pairwise import linear_kernel

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

영화 제목을 입력으로 받아 가장 유사한 영화 10편의 목록을 출력하는 함수를 만들것이다.

이 함수를 만들기 위해 우리는 영화의 제목과 dataframe의 index를 reverse mapping해야한다.

```python
indices = pd.Series(df2.index, index=df2['title']).drop_duplicates()
indices
```

steps
- 제목으로 주어진 영화의 인덱스를 가져온다
- Get the list of cosine similarity scores for that particular movie with all movies. Convert it into a list of tuples where the first element is its position and the second is the similarity
- Sort the aforementioned list of tuples based on the similarity scores; that is, the second element
- Get the top 10 elemennts of this list. ignore the first element as it refers to self
- Return the titles corresponding to the indices of the top elements

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = indices[title]
    
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    sim_scores = sim_scores[1:11]
    
    movie_indices = [i[0] for i in sim_scores]
    
    return df2['title'].iloc[movie_indices]
```

```python
get_recommendations('The Dark Knight Rises')
```

Quality of recommendations is not great. 'The Dark Knight Rises' returns only Batman movies 

## Credits, Genres and Keywords Based Recommender

We're going to build a recommender based on the following metadata: the 3 top actors, related genres and the movie plot keywords.

from the cast, crew, keywords features, we need to extract the three most important actors, the director and the keywords associated with that movie

```python
from ast import literal_eval

features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(literal_eval)
```

Next, make function that can help to extract the required information from each feature

```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        
        if len(names) > 3:
            names = names[:3]
        return names
    
    return []
```

```python
df2['director'] = df2['crew'].apply(get_director)

features = ['cast','keywords','genres']
for feature in features:
    df2[feature] = df2[feature].apply(get_list)
```

```python
df2[['title','cast','director','keywords','genres']].head(3)
```

next, would be to convert the names and keyword instances into lowercase and strip all the spaces between them. this is done so that our vectorizer dosen't count the same word

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
features = ['cast', 'keywords', 'director', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(clean_data)
```

Now we can create "metadata soup", which is a string that contains all the metadata that we want to feed to our vectorizer

```python
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
df2['soup'] = df2.apply(create_soup, axis=1)
```

The next steps are smae as what we did with our plot description based recommneder, but one important difference is we use **CountVectorizer()** isntead of TF-IDF

because we don't want to down-weight the presence of an actor/director if he or she acted or directed in relatively more movies. It doesn't make much intuitive sense.

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df2['soup'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
df2 = df2.reset_index()
indices = pd.Series(df2.index, index=df2['title'])
```

```python
get_recommendations('The Dark Knight Rises', cosine_sim2)
```

Now can see our recommender has been successful in capturing more information due to more metadata and has given us better recommendation

# Collaborative Filtering

우리의 콘텐츠기반 엔진은 특정영화에 비슷한 것 밖에 추천하지 못했다. 즉 장르, 취향에 맞게 추천하는 능력이 없었다. 

따라서 Collaborative Filtering을 사용한다. 
이 시스템은 유저에게 본인과 비슷한 유저가 좋아하는 제품을 추천한다. 두 유저간의 유사도 측정을 위해 우리는 pearson correlation 또는 cosine similarity를 이용할 수 있다. 

예를 들어 아래 테이블의 각 row는 user를 나타낸다, 마지막 column, 즉 해당 사용자와 타켓 사용자 간의 유사성을 기록하는 column을 제외하고 column들은 각 영화에 대응한다. 각 셀은 사용자가 해당 동영상에 부여하는 등급을 나타낸다. 



유저 E가 타켓이라고 가정한다. 유저 A와 유저 F는 사용자 E와 겹치게 영화 평점을 메기지 않으므로 Pearson Correlation에선 그들의 유사도가 정의되지 않는다. 그러므로 우리는 유저 B,C,D만 고려하면 된다. 아래는 유사도를 계산해준 테이블이다.



위의 테이블에서 유저 D가 유저 E와 매우 다르다는 것을 알 수 있다. Pearson Correlation이 음수이기 때문이다. 이제 유저 E가 점수를 주지 않은 영화의 빈칸을 다른 유저를 기반으로 채운다.



user-based CF를 계산하는 것은 쉽지만, 문제점을 동반한다. 그 중 한가지는 유저의 선호도가 시간이 지남에 따라 변할 수 있다는 것이다. 이는 주변 사용자를 기반으로 테이블을 미리 계산하면 성능이 저하될 수 있음을 나타낸다. 문제를 해결을 위해 item-based CF를 이용한다.

## Item Based Collaborative Filtering

유저간의 유사도를 측정하는 대신 item-based CF는 타겟 사용자가 평가한 item과의 유사성에 따라 item을 추천한다. 마찬가지로 Pearson Correlation 또는 Cosine Similarity로 계산할 수 있다. 주요 차이점으로 user-based CF가 수평으로 빈칸을 채우는 것에 비해, item-based collaborative filtering은 수직으로 빈칸을 채운다. 아래의 테이블은 영화 "Me Before You"에 대한 작업 수행을 보여준다.



item-based CF가 정적이기 때문에 동적인 사용자 선호 문제를 방지한다. 하지만 이 방법에는 몇가지 문제가 더 남아있다. \
첫째로, scalability(확장성)이다. 계산은 고객과 제품의 양에 따라 늘어난다, m(users), n(items)일 경우 시간복잡도가 O(mn)이다. \
둘째로 sparsity(희소성) 문제도 있다. 위의 테이블을 보면 타이타닉과 매트릭스를 둘 다 평가한 유저는 한명이지만, 그들의 유사도는 1이다. 극단적인 경우 수백만 명의 사용자가 있을 수 있고 두 개의 상당히 다른 영화 사이의 유사성이 매우 높을 수 있다. 
왜냐하면 둘 다 순위를 매긴 유일한 사용자에 대해 비슷한 점수를 차지하기 때문이다.

## Single Value Decomposition

scalability와 sparsity 문제를 해결하는 한 가지 방법은 **Latent Factor Model** 활용하여 user와 item간의 유사성을 찾는 것이다.문제를 최적화 하기를 원하므로 RMSE를 이용하여 user가, 지정한 item의 등급을 예측하는데 있어 얼마나 우수한지 알 수 있다. 

그렇다면 Latent Factor Model(잠재 요소 모델)이 뭘까? 이것은 user또는 item이 갖는 속성 또는 개념을 설명하는 광범위한 아이디어이다. 예를 들어 음악의 경우 잠재 요소는 음악이 속한 장르를 가리킬 수 있다.\
SVD는 잠재 요인을 추출하여 효용 행렬의 치수를 감소시킨다. 기본적으로 각 user와 각 item을 r차원의 잠재 공간에 매핑한다. 이로써 user와 item을 직접 비교할 수 있게 되면 user간의 관계를 더 잘 이해할 수 있다.  



우리가 전에 사용했던 데이터셋에는 userid가 없다. 이는 collaborative filtering에서는 필수적이므로 새로운 데이터셋을 로드한다. 

```python
from surprise import Reader, Dataset, SVD
from surprise.model_selection import cross_validate
reader = Reader()
ratings = pd.read_csv(r'D:\다운로드\dataset\movie_rec\ratings_small.csv')
ratings.head()
```

```python
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)
svd = SVD()
```

```python
cross_validate(svd, data, measures=['RMSE','MAE'], cv=5)
```

```python
trainset = data.build_full_trainset()
svd.fit(trainset)
```
