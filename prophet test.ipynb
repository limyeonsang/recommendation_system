```python
import pandas as pd
from fbprophet import Prophet
```

```python
df = pd.read_csv(r'D:\다운로드\kaggle dataset\avocado.csv')
df.head()
```

```python
df.describe()
```

```python
df.groupby('type').mean()
```

```python
df = df.loc[(df.type == 'conventional') & (df.region=='TotalUS')]

df['Date'] = pd.to_datetime(df['Date'])

data = df[['Date', 'AveragePrice']].reset_index(drop=True)

data = data.rename(columns={'Date':'ds', 'AveragePrice':'y'})

data.head()
```

```python
data.plot(x='ds', y='y')
```

```python
model = Prophet()

model.fit(data)
```

```python
future = model.make_future_dataframe(periods=365)

forecast = model.predict(future)

forecast.tail()
```

```python
fig1 = model.plot(forecast) # 검은 점이 없는 부분이 예측
```

```python
fig2 = model.plot_components(forecast) # 예측에 영향을 준 요소를 출력
```
